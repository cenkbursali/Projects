## Introduction

In this project, we will work on a [dataset](http://www.randalolson.com/2014/06/14/percentage-of-bachelors-degrees-conferred-to-women-by-major-1970-2012/) that includes the rates of women at the bachelor's degree. The dataset includes the percentage of women in 17 different departments from 1970 to 2012. These 17 major STEMs are covered under three basic categories.

We are planning to visualize aesthetically and simply the entire data with line plots. We will develop our plots step by step. 

## Importing libraries

```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

## Reading Data

The data is divided into 3 groups using the column names of the dataset.

```python
women_rate = pd.read_csv("percent-bachelors-degrees-women-usa.csv")

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

cats = [stem_cats, lib_arts_cats, other_cats]
```

## Plotting first figures

Plots are drawn for all majors in 6x3 layout.

- The limits of the axes are determined.
- The title is added.
- Tickes are removed
- Spins are made invisible.
- *male* and *female* labels are set.

```python
fig = plt.figure(figsize=(15,25))

for i in range(len(cats)):     
    for n in range(0,len(cats[i])):
        ax = fig.add_subplot(6,3,3*n+i+1)
        ax.plot(women_rate["Year"], women_rate[cats[i][n]], linewidth=3)
        ax.plot(women_rate["Year"], 100-women_rate[cats[i][n]], linewidth=3)

        for key, spine in ax.spines.items():
            spine.set_visible(False)
        ax.set_xlim(1968,2011)
        ax.set_ylim(0,100)
        ax.set_title(cats[i][n])
        ax.tick_params(bottom=False, left=False)

        if n == 0:
            if i==0:
                ax.text(2002, 85, 'Women')
                ax.text(2005, 13, 'Men')
            elif i==1:
                ax.text(2002, 78, 'Women')
                ax.text(2005, 20, 'Men')
            elif i==2:
                ax.text(2002, 92, 'Women')
                ax.text(2005, 5, 'Men')
        elif n == 5:
            if i==0:
                ax.text(2005, 87, 'Men')
                ax.text(2002, 8, 'Women')
            elif i==1:
                ax.text(2005, 58, 'Men')
                ax.text(2002, 42, 'Women')
            elif i==2:
                ax.text(2005, 70, 'Men')
                ax.text(2002, 27, 'Women')

plt.show()
```

While men are more dominant in STEM departments, women are more dominant in library-art and other departments. In some departments, the rate of women is around 20%, in some the opposite of this.

The rate of women increases in many departments in time.
