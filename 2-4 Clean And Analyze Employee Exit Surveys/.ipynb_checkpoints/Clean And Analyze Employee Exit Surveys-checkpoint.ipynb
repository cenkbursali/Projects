## Introduction

In this project, we will work on exit surveys of the Department of Education, Training and Employment (DETE) and the Technical and Further Education (TAFE) institutes in Australia. Datasets can be downloaded from [here](https://data.gov.au/dataset/ds-qld-89970a3b-182b-41ea-aea2-6f9f17b5907e/details?q=exit%20survey) for TAFE and [here](https://data.gov.au/dataset/ds-qld-fe96ff30-d157-4a81-851d-215f2a0fe26d/details?q=exit%20survey) for DETE.

We aim to answer to questions below:
- Was dissatisfaction the reason people who worked at the institute for a short time to quit their jobs? What is the conditon with long-term employees?
- Are young employees leaving because they are not satisfied? What about older ones?

We will combine the two datasets to answer the questions. Although the two institutes have the same survey format, they contain different questions.

## Determining missing values 

There are values entered as `Not Stated` in the DETE dataset. These are essentially missing values. But pandas cannot identify them as missing values in this way. Therefore, we must change `Not Stated` to `NaN` in DETE dataset.

```python
dete_survey = pd.read_csv("data/dete_survey.csv", na_values=["Not Stated"])
dete_survey.head()
```

## Removing columns

There were many irrelevant columns in both data sets for our work. We dropped them from data sets.

```python
dete_survey_updated = dete_survey.drop(dete_survey.columns[28:49], axis=1)
tafe_survey_updated = tafe_survey.drop(tafe_survey.columns[17:66], axis=1)
```

## Renaming columns

Now, we will focus on column names. We will make names of columns contain similar information same. 

```python
dete_survey_updated.columns = dete_survey_updated.columns.str.lower().str.strip().str.replace(" ", "_")

columns = {'Record ID': 'id', 'CESSATION YEAR': 'cease_date', 'Reason for ceasing employment': 'separationtype',
'Gender. What is your Gender?': 'gender', 'CurrentAge. Current Age': 'age',
'Employment Type. Employment Type': 'employment_status', 'Classification. Classification': 'position',
'LengthofServiceOverall. Overall Length of Service at Institute (in years)': 'institute_service',
'LengthofServiceCurrent. Length of Service at current workplace (in years)': 'role_service'}
tafe_survey_updated.rename(columns=columns, inplace=True)

tafe_survey_updated.columns
```

## Analyzing types of leaving

We will work on "separationtype" columns in both data sets. 

```python
print(dete_survey_updated["separationtype"].value_counts())
print("")
print(tafe_survey_updated["separationtype"].value_counts())
```

There are different resignation types in the DETE dataset. We combine them to make it common with the TAFE data set.

```python
dete_survey_updated['separationtype'] = dete_survey_updated['separationtype'].str.split('-').str[0]
dete_survey_updated['separationtype'].value_counts()
```

Since we are concerned with leaving due to resignation, we delete other types of separation.

```python
dete_resignations = dete_survey_updated[dete_survey_updated['separationtype'] == 'Resignation'].copy()
tafe_resignations = tafe_survey_updated[tafe_survey_updated['separationtype'] == 'Resignation'].copy()
```

## Analyzing start and cease dates

Now we focus on employee entry and exit dates.

```python
dete_resignations["cease_date"].value_counts()
```

It seems that the data in the `cease_date` column are not in the same format. We make the dates in this column standard convert them to float.

```python
dete_resignations["cease_date"] = dete_resignations["cease_date"].str.split("/").str.get(-1).astype("float")
dete_resignations['cease_date'].value_counts()
```

We examine the start date of employees and check it for outlier values.

```python
dete_resignations["dete_start_date"].value_counts().sort_values()
```

There appears to be no problem with this column.

We examine the `cease_date` column of the TAFE dataset.

```python
tafe_resignations['cease_date'].value_counts().sort_values()
```

## Getting service duration of employees 

In order to achieve our ultimate goal in this project, we need service time of employees at the institute. This information is available in TAFE dataset but not in DETE dataset. That's why we need to produce it by using present columns of DETE dataset.

```python
dete_resignations["institute_service"] = dete_resignations["cease_date"]-dete_resignations["dete_start_date"]
dete_resignations["institute_service"].head()
```

## Finding out dissatisfied employees

We have lists of employees who resigned from the two institutes. However, we have no idea why they quit. We focus on this information now. We will use columns below for both datasets.

tafe_survey_updated:
- `Contributing Factors. Dissatisfaction`
- `Contributing Factors. Job Dissatisfaction`

dete_survey_updated:
- `job_dissatisfaction`
- `dissatisfaction_with_the_department`
- `physical_work_environment`
- `lack_of_recognition`
- `lack_of_job_security`
- `work_location`
- `employment_conditions`
- `work_life_balance`
- `workload`

```python
print(tafe_resignations["Contributing Factors. Dissatisfaction"].value_counts())
print()
print(tafe_resignations["Contributing Factors. Job Dissatisfaction"].value_counts())
```

```python
def update_vals(val):
    if val == "-":
        return False
    elif pd.isnull(val):
        return np.nan
    else:
        return True 
  
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']].applymap(update_vals).any(1, skipna=False)
tafe_resignations_up = tafe_resignations.copy()

tafe_resignations_up['dissatisfied'].value_counts(dropna=False)
```

```python
dete_resignations["dissatisfied"] = dete_survey_updated[["job_dissatisfaction", "dissatisfaction_with_the_department",
                   "physical_work_environment", "lack_of_recognition",
                   "lack_of_job_security", "work_location", 
                   "employment_conditions", "work_life_balance", "workload"]].any(
axis=1, skipna=False)
dete_resignations_up = dete_resignations.copy()
dete_resignations_up['dissatisfied'].value_counts(dropna=False)
```

## Combining datasets

When we combine the datasets directly, we can not follow which row belongs to which institute. We create a new column in both datasets to prevent this confusion.

```python
dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'
```

```python
combined = pd.concat([dete_resignations_up, tafe_resignations_up], ignore_index=True)
combined.notnull().sum().sort_values()
```

```python
combined_updated = combined.dropna(thresh = 500, axis =1).copy()
```

## Reorganizing `institute_service` column

We have to reorganize `institute_service` column. Values in this column can be seen below. 

```python
combined_updated["institute_service"].value_counts(dropna=False)
```

Values in this column are not in the same format as seen. Some values are time range, some are year. They have to be in same data type. 

```python
combined_updated['institute_service_up'] = combined_updated['institute_service'].astype('str').str.extract(r'(\d+)')
combined_updated['institute_service_up'] = combined_updated['institute_service_up'].astype('float')

combined_updated['institute_service_up'].value_counts()
```

Now, service duration can be categorized. 

```python
def transform_service(val):
    if val >= 11:
        return "Veteran"
    elif 7 <= val < 11:
        return "Established"
    elif 3 <= val < 7:
        return "Experienced"
    elif pd.isnull(val):
        return np.nan
    else:
        return "New"
combined_updated['service_cat'] = combined_updated['institute_service_up'].apply(transform_service)

# Quick check of the update
combined_updated['service_cat'].value_counts()
```

## Analyzing all data

Finally, our data is ready for seeking answers to questions below. 

- Was dissatisfaction the reason people who worked at the institute for a short time to quit their jobs? What is the conditon with long-term employees?
- Are young employees leaving because they are not satisfied? What about older ones?

```python
combined_updated['dissatisfied'].value_counts(dropna=False)
```

We have still missing data in `dissatisfied` column. Let's fill them with `False`. 

```python
combined_updated['dissatisfied'] = combined_updated['dissatisfied'].fillna(False)
```

Let's calculate the percentage of employees who resigned because they are dissatisfied.

```python
dis_pct = combined_updated.pivot_table(index='service_cat', values='dissatisfied')

%matplotlib inline
dis_pct.plot(kind='bar', rot=30)
```
